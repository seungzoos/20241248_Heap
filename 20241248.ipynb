### 3번 문제

```python
class Heap:
    def __init__(self, *args):
        if len(args) != 0:
            self.__A = args[0]  
        else:
            self.__A = []
    def insert(self, x):
        self.__A.append(x)
        self.__percolateUp(len(self.__A) - 1)

    def __percolateUp(self, i: int):
        parent = (i - 1) // 2
        if i > 0 and self.__A[i] > self.__A[parent]:
            self.__A[i], self.__A[parent] = self.__A[parent], self.__A[i]
            self.__percolateUp(parent)

    def deleteMax(self):
        if not self.isEmpty():
            max = self.__A[0]
            self.__A[0] = self.__A.pop()  
            self.__percolateDown(0)
            return max
        else:
            return None

    def __percolateDown(self, i: int):
        child = 2 * i + 1  
        right = 2 * i + 2  
        if child <= len(self.__A) - 1:
            if (right <= len(self.__A) - 1 and self.__A[child] < self.__A[right]):
                child = right  # index of larger child
            if self.__A[i] < self.__A[child]:
                self.__A[i], self.__A[child] = self.__A[child], self.__A[i]
                self.__percolateDown(child)

    def max(self):
        return self.__A[0]

    def buildHeap(self):
        for i in range((len(self.__A) - 2) // 2, -1, -1):
            self.__percolateDown(i)

    def isEmpty(self) -> bool:
        return len(self.__A) == 0

    def clear(self):
        self.__A = []

    def size(self) -> int:
        return len(self.__A)
```

### 4번 문제

```python
class Node:
    def __init__(self, data):
        self.data = data 
        self.prev = None
        self.next = None

class CircularDoublyLinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)

        if not self.head:
            self.head = new_node
            new_node.next = new_node
            new_node.prev = new_node
        else:
            tail = self.head.prev
            tail.next = new_node
            new_node.prev = tail
            new_node.next = self.head
            self.head.prev = new_node

    def traverse(self):
        result = []
        if not self.head:
            return result

        current = self.head
        while True:
            result.append(current.data)
            current = current.next
            if current == self.head:
                break
        return result
```

```python
import pandas as pd

class Node:
    def __init__(self, data):
        self.data = data  
        self.prev = None
        self.next = None

class CircularDoublyLinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            new_node.next = new_node
            new_node.prev = new_node
        else:
            tail = self.head.prev
            tail.next = new_node
            new_node.prev = tail
            new_node.next = self.head
            self.head.prev = new_node

    def traverse(self):
        result = []
        if not self.head:
            return result
        current = self.head
        while True:
            result.append(current.data)
            current = current.next
            if current == self.head:
                break
        return result

my_team = {'선예영', '황채원', '나현주', '윤서현', '이승주', 
           '조승연', '남수연', '김민선', '김하린', '이서진'}

csv_path = r'C:\학교\2025-01\자료구조\과제\birthday.csv'

df = pd.read_csv(csv_path)

birth_col = '생년월일8자리(예.20040101)'
df[birth_col] = pd.to_datetime(df[birth_col], format="%Y%m%d", errors='coerce')

cdll = CircularDoublyLinkedList()

for _, row in df.iterrows():
    cdll.append((row['이름'], row[birth_col]))

print("같은 조 친구들의 생일:")
for name, birthday in cdll.traverse():
    if name in my_team:
        if pd.notna(birthday): 
            print(f"{name}: {birthday.strftime('%Y-%m-%d')}")
        else:
            print(f"{name}: 생일 정보 없음")
```

### 5. 교재 8장 우선 순위 큐 연습문제

-> 최대 힙에서는 부모 노드가 자식 노드보다 크거나 같다. 그래서 루트부터 아래로 갈수록 값이 작아진다. 따라서 더 얕은 곳에 있는 원소가 더 작은 값을 가질 수 없다.

2번
-> 아니다. 최대 힙에서 A[0]은 항상 가장 큰 값을 가지지만, A[n-1]은 그렇지 않다. A[n-1]은 단순히 마지막에 삽입된 원소이고, 이것이 가장 작은 값이라는 보장은 없다.

3번
-> [n/2]개이다.
buildHeap() 알고리즘에서는 리프 노드는 자식이 없으므로 스며내리기(sift down)를 할 필요가 없다.
n개의 원소 중 절반 정도는 리프 노드이므로, 확인만 하고 넘어간다.

4번
-> 최악의 경우에 해당하는 시간은 Θ(log n)이고, 최선의 경우에 하당하는 시간은 Θ(1)이다.

5번
-> 간단한 일이다.
힙의 마지막 원소는 리프 노드이자 배열의 끝에 있으므로, 단순히 그 원소를 제거하면 된다.
스며내리기나 올려올리기 같은 힙 속성 유지 작업은 필요 없다. 따라서 테스트 목적의 삭제라면 쉽게 처리할 수 있다.

6번
-> 아니다. 위쪽에서부터 스며오르기를 반복하면 힙은 만들어지지만, 점근적으로 더 비효율적이다.
아래쪽에서 스며내리기를 하는 buildHeap() 알고리즘은 Θ(n) 시간에 힙을 만들 수 있다.
반면, 위쪽에서 스며오르기를 하면 각 삽입마다 Θ(log n)이 걸려서 전체적으로 Θ(n log n) 시간이 든다.
따라서, 본문에서 제시한 방법(스며내리기 방식)이 더 효율적이다.

7번
-> 최대 힙에서 어떤 원소의 값이 증가하면, 해당 원소가 부모 노드보다 더 커질 수 있기 때문에 힙의 성질을 위반하게 된다. 이 경우에는 스며오르기(sift up) 과정을 통해 힙을 수선할 수 있다. 먼저 값이 증가한 노드에서 시작해 부모 노드와 값을 비교하고, 만약 부모보다 크다면 두 값을 교환한다. 이 과정을 루트 노드에 도달하거나 더 이상 부모보다 크지 않을 때까지 반복한다. 이렇게 하면 최대 힙의 성질을 유지하면서 변경된 값을 반영할 수 있고, 이 과정은 힙의 높이에 비례하므로 시간 복잡도는 O(log n)이다.


### 6. LeetCode 703.Kth Largest Element in Stream 문제

```python
import heapq

class KthLargest:

    def __init__(self, k: int, nums: list[int]):
        self.k = k
        self.min_heap = nums
        heapq.heapify(self.min_heap)  

        while len(self.min_heap) > k:
            heapq.heappop(self.min_heap)

    def add(self, val: int) -> int:
        heapq.heappush(self.min_heap, val)  

        if len(self.min_heap) > self.k:
            heapq.heappop(self.min_heap)  

        return self.min_heap[0]  
```
